# Data cleaning in Pandas

```python
import pandas as pd 
```

Import data

```python
df = pd.read_excel('Customer Call List.xlsx')
```

Suppression des doublons dans  le dataset en utilisant **drop_duplicates**
Qui Par défaut, supprime les lignes en double en fonction de toutes les colonnes.

```python
df = df.drop_duplicates()
df
```

Suppression de la colonne Not_Useful_Columun:

Dans notre jeu de données, nous remarquons que nous aurons pas besoins de la colonne Not_Useful_Columun en utilisant la fonction **drop**

```python
df = df.drop(columns= "Not_Useful_Column")
df
```

En examinant notre dataset, nous remarquons que certains elements dans la colonne Last_Name contiennent des caracteres speciaux, ce que nous devons retirer. 
Pour le faire, nous allons utiliser la fonction **str.strip**

Series.str.strip
Supprimez les caractères de début et de fin dans Série/Index.

Series.str.lstrip
Supprimez les caractères principaux dans Série/Index.

Series.str.rstrip
Supprimez les caractères de fin dans Série/Index.

```python
#df['Last_Name'] = df['Last_Name'].str.lstrip("...")
#df['Last_Name'] = df['Last_Name'].str.lstrip("/")
#df['Last_Name'] = df['Last_Name'].str.strip("_")

df['Last_Name'] = df['Last_Name'].str.strip("123._/")
df
```
